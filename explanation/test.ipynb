```python
import numpy as np
import scipy.sparse as sp
from src.pagerank.core import PageRankEngine

# Create a small 3-node cycle graph: 0->1->2->0
# Each node has out-degree 1, so the transition matrix A is:
# A[1,0] = 1.0 (from 0 to 1)
# A[2,1] = 1.0 (from 1 to 2)  
# A[0,2] = 1.0 (from 2 to 0)
N = 3
rows = [1, 2, 1]  # destination nodes
cols = [0, 1, 1]  # source nodes
data = [1.0, 1.0, 1.0]  # transition probabilities
A = sp.csr_matrix((data, (rows, cols)), shape=(N, N))
print(A)
```

```python
# No dangling nodes (all nodes have out-degree 1)
dangling_mask = np.array([0,0,1])#np.zeros(N, dtype=bool) 
dangling_mask
```

```python
# Create PageRank engine with default parameters (p=0.15, tol=1e-8, max_iter=200)
engine = PageRankEngine(A, dangling_mask)
engine
```

```python
# Run power iteration to compute PageRank
result = engine.run()

# Print results
print(f"Converged: {result.converged}")
print(f"Iterations: {result.iterations}")
print(f"Elapsed time: {result.elapsed_sec:.6f} seconds")
print(f"PageRank scores: {result.scores}")
print(f"Sum of scores: {result.scores.sum():.10f}")  # Should be 1.0
print(f"Final residual: {result.residuals[-1]:.2e}")

# Get top-3 nodes (all should be equal in this symmetric graph)
top_nodes = result.top_k(3)
print("\nTop PageRank nodes:")
for rank, (node_id, score) in enumerate(top_nodes, 1):
    print(f"  {rank}. Node {node_id}: {score:.6f}")
```
